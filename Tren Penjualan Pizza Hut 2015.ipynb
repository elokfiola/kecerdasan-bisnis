```python
#Memasukan data pizza
data = pd.read_excel("/content/Pizza_Sale.xlsx")
data.head(10)
```

```python
print(data.shape)
print(data.info)
```

```python
# Total missing values per kolom
print(data.isnull().sum())

# Persentase missing values per kolom
print((data.isnull().sum() / len(data)) * 100)
```

```python
data = data.dropna()
```

```python
# Total missing values per kolom
print(data.isnull().sum())

# Persentase missing values per kolom
print((data.isnull().sum() / len(data)) * 100)
```

```python
data.shape
```

```python
import pandas as pd

# Pastikan 'order_date' bertipe datetime
data['order_date'] = pd.to_datetime(data['order_date'])

# Format ulang jadi "DD-MM-YYYY"
data['order_date'] = data['order_date'].dt.strftime('%d-%m-%Y')

# Cek hasil
print(data['order_date'].head())
```
